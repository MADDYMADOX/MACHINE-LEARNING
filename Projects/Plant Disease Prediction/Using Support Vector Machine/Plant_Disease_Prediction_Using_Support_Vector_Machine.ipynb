```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
plant_data=pd.read_csv('/content/plant_disease_dataset.csv')
```

```python
plant_data.head()
```

```python
#number of rows and columns in this dataset
plant_data.shape
```

```python
#getting the statistical measures of the dataset
plant_data.describe()
```

```python
plant_data['disease_present'].value_counts()
```

0 --> No Disease

1 --> Disease


```python
plant_data.groupby('disease_present').mean()
```

```python
#separating the data and labels
X= plant_data.drop(columns='disease_present', axis=1)
Y= plant_data['disease_present']
```

```python
print(X)
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standard_data=scaler.transform(X)
```

```python
print(standard_data)
```

```python
X = standard_data
Y = plant_data['disease_present']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine Classifier
classifier.fit(X_train, Y_train)
```

Accuracy Score


Model Evaluation

```python
#accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data :',training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the testing data :',testing_data_accuracy)
```

Making Predictive Model

```python
input_data=(27.712800217929825,48.18594618343214,3.063415847672062,4.42677861557573)

#changing input data into numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

std_data=scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0]==0):
  print('The plant is good')
else:
  print('Plant has Disease')
```
